```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: (x-1)*np.tan(x) + x*np.sin(np.pi * x)
a = np.linspace(0,1,37)

print(f(0), f(1))

plt.plot(a,(f(a)))
plt.axhline(0, color='k')
plt.show()
```

```python
f = lambda x: (x-2)*np.sin(x)*np.log(x+2)
a, b = -1, 3
x = np.linspace(a, b, 37)

print(f(a), f(b))
print(f(0), f(2))

plt.plot(x,(f(x)))
plt.axhline(0, color='k')
plt.show()
```

```python
f = lambda x: (1-np.exp(x)+2*x)/3
a, b = 0,1
x = np.linspace(a, b, 37)

print(f(a), f(b))
print(f(0), f(2))

plt.plot(x,(f(x)))
plt.axhline(0, color='k')
plt.show()
```

```python
f = lambda x: (4*x-3)/(x**2-2)
a, b = 0.05,1.25
x = np.linspace(a, b, 37)

print(f(a), f(b))

plt.plot(x,(f(x)))
plt.axhline(0, color='k')
plt.show()
```
