```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#1.- Demuestre que las siguientes ecuaciones tienen al menos una soluci√≥n en los intervalos dados.

$ (a) \quad x - (\ln(x))^x = 0$, en $[4, 5] $

```python
f = lambda x: x - (np.log(x))**x
a, b = 4, 5
x = np.linspace(a, b, 37)

print(f(a), f(b))

plt.plot(x,(f(x)))
plt.axhline(0, color='k')
plt.show()
```

$ (b) \quad x\cos(x) - 2x^2 + 3x - 1 = 0$, en $[0.2, 0.3]$ y $[1.2, 1.3]$

```python
f = lambda x: x*np.cos(x) -2*x**2 +3*x -1
x = np.linspace(0,1.5,37)

print(f(0.2), f(0.3))
print(f(1.2), f(1.3))

plt.plot(x,(f(x)))
plt.axhline(0, color='k')
plt.show()
```

#3.- Demuestre que las primeras derivadas de las siguientes funciones se anulan al menos una vez en los intervalos dados.

$(a) \quad (x - 1) \tan (x) + x \sin (\pi x)$, en $[0, 1]$

```python
f = lambda x: (x-1)*np.tan(x) + x*np.sin(np.pi * x)
a = np.linspace(0,1,37)

print(f(0), f(1))

plt.plot(a,(f(a)))
plt.axhline(0, color='k')
plt.show()
```

$(b) \quad (x - 2)(\sin (x)) \ln (x + 2)$, en $[‚àí1, 3]$

```python
f = lambda x: (x-2)*np.sin(x)*np.log(x+2)
a, b = -1, 3
x = np.linspace(a, b, 37)

print(f(a), f(b))
print(f(0), f(2))

plt.plot(x,(f(x)))
plt.axhline(0, color='k')
plt.show()
```

$(a) \quad f(x) = \frac{2 - e^{x} + 2x}{3}$, en $[0, 1]$
