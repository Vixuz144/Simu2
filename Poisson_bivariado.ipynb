```python
from random import random
import numpy as np
import matplotlib.pyplot as plt
from numba import jit
```

```python
def bosque(r, ρ, polar = False):
  n = 0
  X = []
  while sum(X) <= np.pi*r**2:
    n += 1
    X.append(-np.log(random())/ρ)

  R, θ = [], []
  i = 1
  while n>1 and i<n:
    R.append(np.sqrt(sum(X[:i])/np.pi))
    θ.append(r*random())
    i += 1

  if polar:
    plt.axes(projection='polar')
    plt.plot(R, θ, '.')
    plt.plot(0,0,'w.')
    plt.title('Proceso de Poisson Bivariado')
    plt.grid(color='gray')
    plt.show()
    return R, θ
  else:
    X = R*np.cos(θ)
    Y = R*np.sin(θ)

    plt.plot(X,Y, '.')
    plt.plot(0,0,'w.')
    plt.title('Proceso de Poisson Bivariado')
    plt.grid(color='gray')
    plt.show()
    return X, Y
```

```python
radio = 100
densidad = 0.005
```

```python
X, Y = bosque(radio, densidad)
```

```python
R, θ = bosque(radio, densidad, True)
```
