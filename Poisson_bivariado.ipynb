```python
from random import random
import numpy as np
import matplotlib.pyplot as plt
from numba import jit
```

```python
def bosque(r, ρ):
  n = 0
  X = []
  while sum(X) <= np.pi*r**2:
    n += 1
    X.append(-np.log(random())/ρ)

  R, U = [], []
  i = 1
  while n>1 and i<n:
    R.append(np.sqrt(sum(X[:i])/np.pi))
    U.append(r*random())
    i += 1
  X = R*np.cos(U)
  Y = R*np.sin(U)
  return X, Y
```

```python
radio = 200
densidad = 0.005
X, Y = bosque(radio, densidad)
# print(X)
# print(Y)
```

```python
plt.plot(X,Y, '.')
plt.plot(0,0,'w.')
a, b = -radio, radio
plt.hlines(a, a, b, 'r')
plt.hlines(b, a, b, 'r')
plt.vlines(a, a, b, 'r')
plt.vlines(b, a, b, 'r')
plt.title('Proceso de Poisson Bivariado')
plt.grid(color='gray')
plt.show()
```
